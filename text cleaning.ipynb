```python
import string
```

```python
import re
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
#load raw text
filename='metamorphosis.txt'
with open(filename , 'r') as file:
    text=file.read()
```

```python
#split into tokens
tokens=word_tokenize(text)
```

```python
#convert to lower case
tokens=[word.lower() for word in tokens]
```

```python
#remove puncutuation marks and replace it with empty spaces...
re_punc= re.compile('[%s]' %re.escape(string.punctuation))
tokens=[re_punc.sub("" , word) for word in tokens]
```

```python
#remove everything that is not alphanumeric this removes the empty spaces also...
tokens = [word for word in tokens if word.isalpha()]
```

```python
# filter out the stop words
stop_words=set(stopwords.words('english'))
```

```python
words=[word for word in tokens if not word in stop_words]
print(words[:100])
```

Some applications would require us to reduce the c=vocabulary size by taking the words to their root form like fisher , fishing , fished to fish , t would be useful in cases like document classification where deeper meaning is not required but  a sense or focus on the sentiment is all that is needed.

```python
from nltk.stem.porter import PorterStemmer
porter=PorterStemmer()
stemmed_tokens=[porter.stem(word) for word in words]
```

```python
print(stemmed_tokens[:50])
```

Preparing the text data with the scikit model


```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer=CountVectorizer()
vectorizer.fit(tokens)
print(vectorizer.vocabulary_)
vector=vectorizer.transform(tokens)
print(vector.shape)
print(type(vector))
print(vector.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfvectorizer=TfidfVectorizer()
tfvectorizer.fit(tokens)
#print(tfvectorizer.vocabulary_)
print(tfvectorizer.idf_)
vector=vectorizer.transform([tokens[0]])
print(vector.shape)
print(vector.toarray())
```
