```
#import libraries
import nltk
from nltk.corpus import names
nltk.download('names')
```

```
from sklearn.model_selection import train_test_split 
```

```
def feature_selection(name):
  return { 'last_name' : name[-1]}
```

```
#create a dataset
boys_names = names.words('male.txt')
```

```
girls_names = names.words('female.txt')
```

```
boys_dataset =[(feature_selection(name),'boy') for name in boys_names]
```

```
girls_dataset =[(feature_selection(name) , 'girl') for name in girls_names]
```

```
data = boys_dataset + girls_dataset
```

```
X,y = list(zip(*data))
```

```
import numpy as np
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size =0.25 , shuffle= True)
#y_train= np.asarray(y_train)
```

```
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_extraction import DictVectorizer
dict_vector = DictVectorizer()
name_classifier = DecisionTreeClassifier()
```

```
# fit the data
dict_vector.fit(X_train)
```

```
# Classify the data
X_train_vectorized = dict_vector.transform(X_train)
```

```
name_classifier.fit(X_train_vectorized , y_train)
```

```
X_test_vectorizer = dict_vector.transform(X_test)
```
